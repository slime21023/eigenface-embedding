```python
# prepare resized img

img_slice=(slice(70, 195), slice(78, 172))

resized_path = "../data/prepare/resized"
if not os.path.exists(resized_path):
    os.mkdir(resized_path)

prepare_path = '../data/prepare/'
dirs = os.listdir(prepare_path)
for dir in dirs:
    n_path = os.path.join(prepare_path, dir)
    if not os.path.isfile(n_path):
        continue
    
    sample = cv.imread(n_path)
    sample_rgb = cv.cvtColor(sample, cv.COLOR_BGR2RGB)
    resize_img = cv.resize(sample_rgb[img_slice[0], img_slice[1]], (90, 120), interpolation=cv.INTER_CUBIC)
    img = Image.fromarray(resize_img)
    img.save(os.path.join(resized_path, dir))

```
